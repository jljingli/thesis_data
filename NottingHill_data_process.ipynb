```python
from pyecharts import options as opts
from pyecharts.charts import Graph,Tab,Scatter
from pyecharts.globals import ThemeType
import openpyxl
from pyecharts.commons.utils import JsCode
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the active sheet

sheet=openpyxl.load_workbook("V4_nh.xlsx").active
```

```python
# Initializing variables

links=[]
color={"EA":"#fe64a3","A":"#f28500","T":"#008080"}
curve={"EA":0.2,"A":0.5,"T":0.8}

# Counting occurrence of values in column 7

rates={}
for i in range(2,sheet.max_row+1):
    data=sheet.cell(row=i,column=7).value
    if data != None:
        for j in data.replace("\n",",").split(","):
            rates[j] = rates.get(j, 0) + 1  
print(rates)
```

```python
# Generating links for the graph

for i in range(2,sheet.max_row+1):
    data=sheet.cell(row=i,column=7).value
    if data != None:
        for j in data.replace("\n",",").split(","):
            value=j.split("-")[1]
            if ("William" in j) and ("Anna" in j):
                print(j)
                links.append({'source': j.split("-")[0], 'target': j.split("-")[2], 'value': j.split("-")[1],
                             "lineStyle":{"color":color[value],"width":rates[j],"curveness":curve[value]}})
            else:
                links.append({'source': j.split("-")[0], 'target': j.split("-")[2], 'value': j.split("-")[1],
                              "lineStyle": {"color": color[value], "width": rates[j], "curveness": curve[value]}})

print(links)
```

```python
# Initializing a dictionary to store scatter data
scatter={}
# Initializing a list to store x-axis values
x=[]

# Iterating through rows to extract scatter data and x-axis values

for i in range(2,56):
    try:
        x.append(str(sheet.cell(row=i,column=3).value))  # Appending column 3 value to x-axis values
        x.append(str(sheet.cell(row=i,column=4).value))  # Appending column 4 value to x-axis values
        text=sheet.cell(row=i,column=6).value            # Extracting text data from column 6
        for name in text.split(","):                     # Splitting text by comma and iterating through each name
            name=name.strip()                            # Removing leading/trailing spaces from the name
            scatter[name] = scatter.get(name, 0) + 1 # Incrementing count for each name in the scatter dictionary
    except:
        continue                                    # Skipping the iteration if an exception occurs (e.g., empty cell)

scatter_event=list(scatter.items())                 # Converting scatter dictionary to a list of key-value pairs
scatter_event.sort(key=lambda x: x[1], reverse=True)  
                        # Sorting the list in descending order based on the frequency (second element in each pair)

# Printing the sorted scatter data    
print(scatter_event)
print("\n")
```

```python
# Defining the nodes for the graph

nodes = [
    {'name': 'William', 'symbolSize': 85,"category":20},
    {'name': 'Spike', 'symbolSize': 16,"category":19},
    {'name': 'Anna', 'symbolSize': 62,"category":3},
    {'name': 'Journalist1', 'symbolSize': 10,"category":26},
    {'name': 'Karen', 'symbolSize': 12,"category":18},
    {'name': 'Bella', 'symbolSize': 20, "category": 13},
    {'name': 'Max', 'symbolSize': 29,"category":12},
    {'name': 'Honey', 'symbolSize': 17, "category": 15},
    {'name': 'Bernie', 'symbolSize': 16, "category": 7},
    {'name': 'LoudMen', 'symbolSize': 8, "category": 5},
    {'name': 'Jeff', 'symbolSize': 8, "category": 14},
    {'name': 'Girl1', 'symbolSize': 9, "category": 11},
    {'name': 'Girl2', 'symbolSize': 8, "category": 21},
    {'name': 'Girl3', 'symbolSize': 9, "category": 10},
    {'name': 'paparazzi', 'symbolSize': 9, "category": 24},
    {'name': 'CoStar', 'symbolSize': 7,"category":8},
    {'name': 'Security', 'symbolSize': 7, "category": 9},
    {'name': 'Tones', 'symbolSize': 7, "category": 6},
    {'name': 'Concierge', 'symbolSize': 12, "category": 1},
    {'name': 'RitzGuest', 'symbolSize': 7, "category": 22},
    {'name': 'Agent', 'symbolSize': 13, "category": 2},
    {'name': 'Journalist2', 'symbolSize': 12, "category": 17},
    {'name': 'Journalist3', 'symbolSize':8, "category": 16},
    {'name': 'Journalist4', 'symbolSize': 7, "category": 23},
    {'name': 'Journalists', 'symbolSize': 7, "category": 27},
    {'name': 'Mom', 'symbolSize': 7, "category": 25},
    {'name': 'Martin', 'symbolSize': 7, "category": 4},

]
```

```python
# Creating a graph instance

G = Graph(init_opts=opts.InitOpts(width='1600px', height='800px',theme=ThemeType.LIGHT))

# Adding graph data and configurations

G.add(
    series_name='',
    nodes=nodes,
    links=links,
    repulsion=2000,
    edge_label=opts.LabelOpts(is_show=True, position='middle', formatter=' '),
    categories=[{
        "name":cat
    }for cat in range(28)],
    layout="circular",
    is_rotate_label=True,
    label_opts=opts.LabelOpts(position='right'),
    linestyle_opts=opts.LineStyleOpts(color='source', curve=0.3, opacity=0.7)
)

# Setting global options for the graph

G.set_global_opts(
    title_opts=opts.TitleOpts(title='Character Graphic'),
    legend_opts=opts.LegendOpts(orient='vertical', pos_left='2%', pos_top='20%')
)

# Creating a scatter plot instance
c = (
    Scatter(init_opts=opts.InitOpts(width="1600px", height="800px", theme=ThemeType.LIGHT))
)

# Adding x-axis data
c.add_xaxis(x)

# Extracting character data and plotting on the scatter plot
character_list = [i[0] for i in scatter_event[:10]]
for character in character_list:
    current_character = []
    for i in range(2, 56):
        people = sheet.cell(row=i, column=6).value
        if people is not None and character in people:
            current_character.append(float(character_list.index(character) + 1))
            current_character.append(float(character_list.index(character) + 1))
        else:
            current_character.append(None)
            current_character.append(None)
    c.add_yaxis(character, current_character, symbol_size=5, label_opts=opts.LabelOpts(is_show=False))
    print("{}:\n".format(character), current_character)
    print("\n")
print(x)
```

```python
# Setting global options for the scatter plot
c.set_global_opts(
    tooltip_opts=opts.TooltipOpts(trigger="axis"),
    toolbox_opts=opts.ToolboxOpts(
        is_show=True,
        pos_top='30px',
        pos_left='400px',
        feature={
            "saveAsImage": {},
            "restore": {},
            "magicType": {
                "show": True,
                "type": ["line", "bar", "stack"]
            },
            "dataView": {}
        }
    ),
    xaxis_opts=opts.AxisOpts(
        type_="category",
        boundary_gap=True,
        axisline_opts=opts.AxisLineOpts(
            linestyle_opts=opts.LineStyleOpts(width=1)
        ),
        splitline_opts=opts.SplitLineOpts(
            is_show=True,
            linestyle_opts=opts.LineStyleOpts(type_="dashed")
        )
    ),
    title_opts=opts.TitleOpts(title="Character-Timestamp"),
    yaxis_opts=opts.AxisOpts(
        axislabel_opts=opts.LabelOpts(
            formatter=JsCode(
                '''
                function(value) {
                    if (value == '1') {
                        return 'William';
                    } else if (value == '2') {
                        return 'Anna';
                    } else if (value == '0') {
                        return ' ';
                    } else if (value == '3') {
                        return 'Max';
                    } else if (value == '4') {
                        return 'Bella';
                    } else if (value == '5') {
                        return 'Honey';
                    } else if (value == '6') {
                        return 'Spike';
                    } else if (value == '7') {
                        return 'Bernie';
                    } else if (value == '8') {
                        return 'Agent';
                    } else if (value == '9') {
                        return 'Journalists';
                    } else {
                        return 'Tones';
                    }
                }
                '''
            )
        ),
        interval=1,
        axisline_opts=opts.AxisLineOpts(
            linestyle_opts=opts.LineStyleOpts(width=1)
        ),
        splitline_opts=opts.SplitLineOpts(
            is_show=True,
            linestyle_opts=opts.LineStyleOpts(type_="dashed")
        )
    )
)


# Creating a tab layout
tab=Tab()
tab.add(G,"Character Graphic")
tab.add(c,"Character Timing")
tab.render("result_final.html")
```

```python
word_counts = {}
for i in range(2, sheet.max_row + 1):
    data = sheet.cell(row=i, column=7).value
    if data is not None:
        words = data.replace("\n", ",").split(",")
        for word in words:
            word = word.strip()
            if word != "":
                if "-" in word:
                    # Split word at the hyphen and count both parts
                    word_parts = word.split("-")
                    for part in word_parts:
                        part = part.strip()
                        if part != "":
                            word_counts[part] = word_counts.get(part, 0) + 1
                else:
                    word_counts[word] = word_counts.get(word, 0) + 1

print("Word Counts:")
for word, count in word_counts.items():
    print(word, ":", count)
```
