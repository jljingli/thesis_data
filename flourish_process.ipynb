```python
# Initialize a dictionary to store the percentage for each relationship value
percentages = {}

# Iterate over each unique event type
for event_type in merged_df["Event Type"].unique():
    # Filter the DataFrame for the current event type
    event_df = merged_df[merged_df["Event Type"] == event_type]
    
    # Calculate the count of each relationship value for the current event type
    relationship_counts = event_df["ActionType"].value_counts()

    # Calculate the percentage for each relationship value
    event_percentages = relationship_counts / relationship_counts.sum() * 100
    
    # Store the percentages in the dictionary
    percentages[event_type] = event_percentages

# Print the percentages for each event type
for event_type, percentage in percentages.items():
    print(f"Event Type: {event_type}")
    print(percentage)
    print()
```

```python
# Initialize a dictionary to store the percentage for each relationship value
percentages = {}

# Iterate over each unique event type
for event_type in merged_df["Event Type"].unique():
    # Filter the DataFrame for the current event type
    event_df = merged_df[merged_df["Event Type"] == event_type]
    
    # Create an empty dictionary to store the counts for each relationship value
    relationship_counts = {}
    
    # Iterate over each row in the current event DataFrame
    for row in event_df.itertuples(index=False):
        # Split the ActionType values using semicolon as the delimiter
        action_types = row.ActionType.split(",")
        
        # Update the counts for each relationship value
        for action_type in action_types:
            if action_type in relationship_counts:
                relationship_counts[action_type] += 1
            else:
                relationship_counts[action_type] = 1
    
    # Calculate the total count of relationship values
    total_count = sum(relationship_counts.values())

    # Calculate the percentage for each relationship value
    event_percentages = {action_type: count / total_count * 100 for action_type, count in relationship_counts.items()}
    
    # Store the percentages in the dictionary
    percentages[event_type] = event_percentages

# Print the percentages for each event type
for event_type, percentage in percentages.items():
    print(f"Event Type: {event_type}")
    print(percentage)
    print()
```

```python
from IPython.display import display
```

```python
# Create a new DataFrame to display the percentages
result_df = pd.DataFrame(percentages)

# Transpose the DataFrame to have event types as columns
result_df = result_df.transpose()

result_df = result_df.applymap(lambda x: "{:.2f}%".format(x))

# Set column names and sort columns alphabetically
result_df = result_df.sort_index(axis=1)

# Display the table
display(result_df)

result_df.to_csv("result_table.csv")
```
